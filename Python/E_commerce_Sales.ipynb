```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Ecommerce_Sales_Data_2024_20251(E-commerce_Sales).csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
esale = df.groupby("Category")["Sales"].sum().sort_values()
plt.barh(esale.index, esale.values)
plt.xlabel('Sales')
plt.ylabel('Category')
plt.title('Sales by Category')
plt.show()
```

```python
df['Category'].value_counts()
```

```python
df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
eprofit = df.groupby('Category')['Profit'].sum().sort_values()
plt.barh(eprofit.index, eprofit.values)
plt.xlabel('Profit')
plt.ylabel('Category')
plt.title('Profit by Category')
plt.show()
```

```python
df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
citys = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
plt.bar(citys.index, citys.values)
plt.xlabel('City')
plt.xticks(rotation=90)
plt.ylabel('Sales')
plt.title('Sales By City')
plt.show()
```

```python
df.groupby('City')['Profit'].sum().sort_values(ascending=False).head(10)
```

```python
cityp = df.groupby('City')['Profit'].sum().sort_values(ascending=False).head(10)
plt.bar(cityp.index, cityp.values)
plt.xlabel('City')
plt.xticks(rotation=90)
plt.ylabel('Sales')
plt.title('Profit By City')
plt.show()
```

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
subC = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
plt.bar(subC.index, subC.values)
plt.xlabel('Sub-Category')
plt.tick_params(axis='x', rotation=90)
plt.ylabel('Sales')
plt.title('Sales By Sub-Category')
plt.show()
```

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)
```

```python
sub_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)
plt.bar(sub_profit.index, sub_profit.values)
plt.xlabel('Sub-Category')
plt.tick_params(axis='x', rotation=90)
plt.ylabel('Profit')
plt.title('Profit By Sub-Category')
plt.show()
```

```python
df['Region'].value_counts()
```

```python
Region = df['Region'].value_counts()
plt.pie(Region.values, labels=Region.index, autopct='%1.1f%%')
plt.title('Region Distribution')
plt.show()
```

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
product_sale = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.bar(product_sale.index, product_sale.values)
plt.xlabel('Product Name')
plt.tick_params(axis='x', rotation=90)
plt.ylabel('Sales')
plt.title('Sales By Product')
plt.show()
```
